[Create a Column Stacked Graph Based On a Pandas' DataFrame](https://www.youtube.com/watch?v=zeX9Uu11lJ8)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(538)
df = pd.DataFrame({
	'a': np.random.randint(0, 5, size=5),
	'b': np.random.randint(0, 5, size=5),
	'c': np.random.randint(0, 5, size=5)
	})

print(df)

df.plot(kind='barh', stacked=True)
df.plot(kind='bar', stacked=True)
plt.legend(loc='upper left')
plt.show()
```

```python
indx = []
indx.append(df['a'][0] + df['b'][0] + df['c'][0])
indx.append(df['a'][1] + df['b'][1] + df['c'][1])
indx.append(df['a'][2] + df['b'][2] + df['c'][2])
indx.append(df['a'][3] + df['b'][3] + df['c'][3])
indx.append(df['a'][4] + df['b'][4] + df['c'][4])

print(indx)
```

```python
max_value =  max(indx)
bar_value = []

for i in range(0, len(indx)):
    bar_value.append(float((indx[i]/max_value)*100))
    
print(bar_value)

df = pd.DataFrame(bar_value)
print(df)
w = df[0].tolist()
print(w)

df.plot(kind='bar', stacked=True, width = .5)
plt.legend(loc='upper left')
plt.show()
```

# [Control bar width and position in barplots](https://python-graph-gallery.com/5-control-width-and-space-in-barplots/)

```python
# library
import matplotlib.pyplot as plt
 
# create dataset
height = [3, 12, 5, 18, 45]
bars = ('AAAAAAAAA', 'B', 'C', 'D', 'E')
 
# Choose the position of each barplots on the x-axis (space=1,4,3,1)
x_pos = [0,1,5,8,9]
 
# Create bars
plt.bar(x_pos, height)
 
# Create names on the x-axis
plt.xticks(x_pos, bars)
 
# Show graphic
plt.show()
```

```python
# library
import matplotlib.pyplot as plt
 
# create dataset
height = [3, 12, 5, 18, 45]
bars = ('AAAAAAAAA', 'B', 'C', 'D', 'E')
 
# Choose the position of each barplots on the x-axis (space=1,4,3,1)
x_pos = [0,1,5,8,9]
 
# Create bars
plt.bar(x_pos, height)
 
# Create names on the x-axis & rotate
plt.xticks(x_pos, bars, rotation = 80)

 
# Show graphic
plt.show()
```

```python
# library
import matplotlib.pyplot as plt
 
# create dataset
height = [3, 12, 5, 18, 45]
bars = ('A', 'B', 'C', 'D', 'E')

# Choose the width of each bar and their positions
width = [0.1,0.2,3,1.5,0.3]
x_pos = [0,0.3,2,4.5,5.5]
 
# Make the plot
plt.bar(x_pos, height, width=width)

# Create names on the x-axis
plt.xticks(x_pos, bars)
 
# Show graphic
plt.show()
```

# [Stacked Histogram](https://www.youtube.com/watch?v=pRfs9tx-OcA)

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('fivethirtyeight')

x = np.random.randn(10000)
plt.hist(x,30, color='red')
plt.show()

y = np.random.randn(10000)
plt.hist(y,30, color='green')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(10000, 3)
plt.hist(x,30, histtype = "barstacked", width = .2)
plt.show()
```
