```python
import pandas as pd
# import numpy as np
# import seaborn as sns
# import matplotlib.pyplot as plt
```

```python
df_dog_breeds=pd.read_csv('A_dog_breeds.csv')
df_dog_breeds.head()
```

```python
df_dog_intell_size=pd.read_csv('A_dog_intell_size.csv')
df_dog_intell_size.head()
```

```python
df_dog_data = pd.merge(df_dog_breeds, df_dog_intell_size, on= "Breed")
df_dog_data.head()
```

```python
df_dog_data.drop(['Height'], axis=1, inplace=True)               # axis=1 is columns    axis = 0 is rows
df_dog_data.head()
```
